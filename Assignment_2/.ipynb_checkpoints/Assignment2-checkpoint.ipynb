# Image Stitching

```python
def warpPerspectivePadded(
        imgl, H, imgr_shape):
    """Performs an affine or Euclidean/rigid warp with padding.
    Parameters
    ----------
    imgl : image to be warped i.e. left image
    imgr : shape of static image i.e. shape of central image
    M : `3x3` Homography matrix.
    Returns
    -------
    src_warped : padded and warped source image
    """
    
    src = imgl
    dst_shape = imgr_shape
    M = H
    
    # it is enough to find where the corners of the image go to find
    # the padding bounds; points in clockwise order from origin
    _, src_w = src.shape[:2]
    lin_pts = np.array([
        [0, src_w],
        [0, 0],
        [1, 1]])
    # transform points
    transf_lin_pts = M.dot(lin_pts) #M[:, :2].dot(lin_pts) + M[:, 2].reshape(3, 1)
    transf_lin_pts /= transf_lin_pts[-1:,:]
    
    # find min and max points
    min_x = np.floor(np.min(transf_lin_pts)).astype(int)
    max_x = np.ceil(np.max(transf_lin_pts)).astype(int)

    # add translation to the transformation matrix to shift to positive values
    anchor_x = 0
    transl_transf = np.eye(3,3)
    if min_x < 0:
        anchor_x = -min_x
        transl_transf[0,2] = anchor_x
        
    shifted_transf = transl_transf.dot(M)
    shifted_transf /= shifted_transf[2,2]

    # create padded destination image
    dst_h, dst_w,_ = dst_shape
    dst_pad_w = anchor_x + max(max_x, dst_w)
    src_warped = cv2.warpPerspective(src, shifted_transf, (dst_pad_w, dst_h), borderMode=cv2.BORDER_CONSTANT, borderValue=0)
    return src_warped
```

#### Find homography using custom RANSAC method and apply warpPerspectivePadded method to left image

```python
def RANSAC(mp1, mp2, N, acc):
   
    max_inliers = 0
    max_l =[]
    max_r = []

    #while(max_inliers < 100):
    for i in range(N):
        rand = random.randint(0,len(mp2)-1)
        l = [mp1[rand]]
        r = [mp2[rand]]
        for i in range(3):
            rand = random.randint(0,len(mp2)-1)
            l.append(mp1[rand])
            r.append(mp2[rand])
        l = np.asarray(l)
        r = np.asarray(r)
        h, __ = cv2.findHomography(l, r, method=0)
        inliers = 0
        for k in range(len(mp1)):
            point = mp2[k]
            newPoint = np.matmul(h, np.append(mp1[k], [1]).transpose())
            diff1 = (point[0]-newPoint[0])
            diff2 = (point[1]-newPoint[1])
            summ = (diff1**2)+(diff2**2)
            distance = math.sqrt(summ)
            if distance < acc:
                inliers += 1
        if inliers > max_inliers:
            max_inliers = inliers
            max_l = l
            max_r = r
            
    print("NUmber of inliners: ", max_inliers)
    
    h, __ = cv2.findHomography(max_l, max_r, method=0)

    mp1i = []
    mp2i = []

    for k in range(len(mp1)):
            point = mp2[k]
            newPoint = np.matmul(h, np.append(mp1[k], [1]).transpose())
            diff1 = (point[0]-newPoint[0])
            diff2 = (point[1]-newPoint[1])
            summ = (diff1**2)+(diff2**2)
            distance = math.sqrt(summ)
            if distance < acc:
                mp1i.append(mp1[k])
                mp2i.append(mp2[k])

    mp1i = np.asarray(mp1i)
    mp2i = np.asarray(mp2i)

    H, __ = cv2.findHomography(mp1i, mp2i, method=0)
    
    return H


# Arrange matching keypoints in two separate lists
GoodMatches = []
for i, m in enumerate(matches):
    if m.distance < 150:
        GoodMatches.append((m.trainIdx, m.queryIdx))

mp2 = np.float32([keypoints_2[i].pt for (i, __) in GoodMatches])
mp1 = np.float32([keypoints_1[i].pt for (__, i) in GoodMatches])

print(len(mp2), len(mp1))

# We shift all the points in the right image over as they are on smaller canvas
#for i in range(len(mp2)):
#    mp2[i][0] += img2.shape[1]

    
#find homography
#M, _ = cv2.findHomography(mp1, mp2, cv2.RANSAC, 10)
M = RANSAC(mp1, mp2, 100, 10)


print(img2.shape)
print(w1+w2)
print(h2)
# Apply transformation to left image
#warped = cv2.warpPerspective(img1, M, (w1 + w2, h2))
warped = warpPerspectivePadded(img1, M, img2.shape)

plt.figure(figsize=(7,7))
plt.imshow(warped)
plt.title("Warped Img1"), plt.axis('off')
plt.show()
```

#### stitch images together using pyramid blend

```python
def pyramid_blend(A, B):
    
    levels = 1
    
    # Generate Gaussian pyramid for A and B
    gp_A = [A.copy()]
    for i in range(levels):
        gp_A.append(cv2.pyrDown(gp_A[-1]))
       
        
    # Generate Gaussian pyramid for B
    gp_B = [B.copy()]
    for i in range(levels):
        gp_B.append(cv2.pyrDown(gp_B[-1]))

    # Generate Laplacian Pyramids for A 
    lp_A  = [gp_A[levels-1]]
    for i in range(levels-1, 0, -1):
        hA, wA, _ = gp_A[i-1].shape
        GE = cv2.pyrUp(gp_A[i], dstsize=(wA, hA))
        L = cv2.subtract(gp_A[i-1], GE)
        lp_A.append(L)
        
        
    # Generate Laplacian Pyramids B
    lp_B  = [gp_B[levels-1]]
    for i in range(levels-1, 0, -1):
        hB, wB, _ = gp_B[i-1].shape
        GE = cv2.pyrUp(gp_B[i], dstsize=(wB, hB))
        L = cv2.subtract(gp_B[i-1], GE)
        lp_B.append(L)

    # Blend images at each level
    LS = []
    for la, lb in zip(lp_A, lp_B):
        ls = np.hstack((la[:, :la.shape[1] - lb.shape[1]], lb))
        LS.append(ls)

    # now reconstruct
    ls_ = LS[0]
    for i in range(1, levels):
        h, w, _ = LS[i].shape
        ls_ = cv2.pyrUp(blend, dstsize=(w, h))
        ls_ = cv2.add(blend, LS[i])

    return ls_.astype(np.uint8)

# Stitch warped hill1 and hill2 using pyramid blending
img_A = warped
img_B = img2
img_12 = pyramid_blend(img_A, img_B)
h12, w12, _ = img_12.shape

# Display hill12
plt.figure(figsize=(20,5))
plt.imshow(img_12)
plt.title('Image 12 panorama with pyramid blending'), plt.axis('off')
plt.show()
```

#### Apply linear blending and display final image

```python
def linear_blend(i_width, r_width, r_index, start, end):
    
    res = np.zeros(i_width).astype(np.float)
    step = (end - start) / (r_width - 1)
    
    idx = r_index - r_width
    for i in range(r_width):
        res[idx] = start + step * i
        idx += 1
    res[idx:] = 1
    return res


# Stitch img12 and warped img3 
#img_A = np.zeros((h12, w12 + w3, 3)).astype(np.uint8)
img_A = img_12
img_B = warped
img_A = cv2.copyMakeBorder(img_A, 0, 0, 0,(img_B.shape[1] - img_A.shape[1]), cv2.BORDER_CONSTANT, value=0)
alpha = linear_blend(img_A.shape[1], 50, round(img_A.shape[1]/2), 0, 1)

# Blend
img_123 = (1-alpha.reshape(1,-1,1)) * img_A + alpha.reshape(1,-1,1) * warped
img_123 = np.uint8(img_123)

h123, w123, _ = img_123.shape

# Display full image
plt.figure(figsize=(20,5))
plt.imshow(img_123)
plt.title('Full panorama after Linear Blending'), plt.axis('off')
plt.show()
```

Pyramid blending causes a prominent line(disconnect to be visible in the image) however with linear blending the edge is more smoothed out.
##### Pyramid Blending Pros:
- no blurry effect caused by images being superimposed
- good for blending photos with a lot of detail because with this method the details can be preserved

##### Pyramid Blending Cons:
- Computationally intensive
- difficult to implement


##### Linear Blending Pros:
- Easy to implement

##### Linear Blening Cons:
- cause blurring effect at the seam of the two images being stitched
- Leads to loss of detail

#### HoG Features

```python
def HoG_features(images, train):
    cell_size = (4,4)
    block_size = (2,2)
    nbins = 9
    
    features = np.zeros((34596,1))
    for img in images:
        
        #create hog object
        hog = cv2.HOGDescriptor(_winSize=(img.shape[1] // cell_size[1] * cell_size[1],
                                      img.shape[0] // cell_size[0] * cell_size[0]),
                                _blockSize=(block_size[1] * cell_size[1],
                                        block_size[0] * cell_size[0]),
                                _blockStride=(cell_size[1], cell_size[0]),
                                _cellSize=(cell_size[1], cell_size[0]),
                                _nbins=nbins)
    
        n_cells = (img.shape[0] // cell_size[0], img.shape[1] // cell_size[1])
    
        # Compute HoG features
        hog_feats = hog.compute(img)
        
        
        #if training, then sum the corresponding pictures for all the pictures
        if (train):
            features += hog_feats
        else:
            features = hog_feats
            
    
    #if training, calculate mean and also create reshaped copy for easier indexing        
    if (train):
        features = features/len(images)
    features_reshaped = features.reshape(n_cells[1] - block_size[1] + 1,
                                         n_cells[0] - block_size[0] + 1,
                                         block_size[1], block_size[0], nbins) \
                                        .transpose((1, 0, 3, 2, 4))  # index blocks by rows first
   
    return features, features_reshaped    
```

#### Function to extract overlapping windows of the image

```python
def sliding_window(image, stepSize, windowSize):
    # slide a window across the image
    for y in range(100, image.shape[0], stepSize):
        for x in range(0, image.shape[1], stepSize):
            # yield the current window
            yield (x, y, image[y:y + windowSize[1], x:x + windowSize[0]])
```

#### Compute eucliden distance between windows from test image and trained features. Highlight window if below threshold.

```python
def findPedestrian(winSize, stride, threshold, img, trained_features, title):
    (winW, winH) = winSize    #window size
    stepSize = stride                #stride
    dim = (128, 128)             #resize dimension

    threshold = threshold            #euclidean distance threshold

    # extract overlapping windows from the test image and process each
    for (x, y, window) in sliding_window(img, stepSize, (winW, winH)):

        # if the window does not meet our desired window size, ignore it
        if window.shape[0] != winH or window.shape[1] != winW:
            continue

        #resize each window
        resized = cv2.resize(window, dim, interpolation = cv2.INTER_AREA)

        #compute HoG features
        features, features_reshaped = HoG_features([resized], False)
        #print("x, y: ", x, y)

        #compute euclidean distance
        diff = trained_features - features
        diff_square = diff ** 2
        euclidean_dist = np.sqrt(np.sum(diff_square))

        #print(euclidean_dist)

        if (euclidean_dist < threshold):
            cv2.rectangle(test_img, (x, y), (x + winW, y + winH), (255, 255, 255), 3)



    # view result
    fig=plt.figure(figsize=(20, 24), dpi= 80)
    plt.subplot(121), plt.imshow(test_img, cmap = 'gray')
    plt.title(title), plt.xticks([]), plt.yticks([])

    plt.show()
    
    return
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Denoising

### 1. Filter using 5x5 Gaussian filter with variance = 2 

```python
# load image
image_path = "./gaussian_lighthouse.png" # path of image to be used for gaussian filtering
img = cv2.imread(image_path)

#filter image with 5x5 gaussian filter with variance 2, or std = sqrt(2)
res1 =  cv2.GaussianBlur(img, (5,5), 1.4142)

# convert from BGR to RGB for displaying
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res1 = cv2.cvtColor(res1, cv2.COLOR_BGR2RGB)

# view image
plt.subplot(121), plt.imshow(img)
plt.title("Original noisy light house"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(res1)
plt.title("5x5 Gaussian"), plt.xticks([]), plt.yticks([])
plt.show()
```

### 2. Filter using 5x5 box filter

```python
#filter image using 5x5 box filter
# create a box filter
kernel = np.ones((5,5), np.float32)/(5*5)

# filter the image
res2 = cv2.filter2D(img, -1, kernel)


#plot
plt.subplot(121), plt.imshow(img)
plt.title("Original noisy light house"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(res2)
plt.title("5x5 box filter"), plt.xticks([]), plt.yticks([])
plt.show()
```

### 3. Compare PSNR of both denoisied images with that of the clean image

```python
# load clean image
image_path_clean = "./lighthouse.png" # path of clean image 
img2 = cv2.imread(image_path_clean)

img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

psnr1 = cv2.PSNR(res1, img2) #psnr of gaussian filtered image
psnr2 = cv2.PSNR(res2, img2) #psnr of box filtered image

print("PSNR of gaussian filtered image : ", psnr1)
print("PSNR of box filtered image : ", psnr2)
```

From the above PSNR values we can see that the gaussian filtered image has a higher PSNR value implying that it is more similiar to the original clean image. We can therfore say that the gaussion filter gives better results.

### 4. Filter salt and pepper noise image with gaussian filter

```python
# load salt and pepper noise image
image_path_2 = "./saltandpepper_lighthouse.png" # path of salt and pepper noise image 
img3 = cv2.imread(image_path_2)

#filter image with 5x5 gaussian filter with variance 2, or std = sqrt(2)
res3 =  cv2.GaussianBlur(img3, (5,5), 1.4142)

# convert from BGR to RGB for displaying
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
res3 = cv2.cvtColor(res3, cv2.COLOR_BGR2RGB)

# view image
plt.subplot(121), plt.imshow(img3)
plt.title("Original salt & pepper lighthouse"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(res3)
plt.title("5x5 Gaussian"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#filter image using a median filter
res4 = cv2.medianBlur(img3, 5)

# view image
plt.subplot(121), plt.imshow(img3)
plt.title("Original salt & pepper lighthouse"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(res4)
plt.title("5x5 median filter"), plt.xticks([]), plt.yticks([])
plt.show()

```

### 6. Compare PSNR of both denoisied images with that of the clean image

```python
psnr3 = cv2.PSNR(res3, img2) #psnr of gaussian filtered image
psnr4 = cv2.PSNR(res4, img2) #psnr of box filtered image

print("PSNR of gaussian filtered image : ", psnr3)
print("PSNR of median filtered image : ", psnr4)
```

The Median filtered image has a higher PSNR value impling that it is more similar to the clean image. Therefore for this application, the median filter gives better results.

# Sobel edge detector

### Sobel edge detector on clean image

Effect of filter size: As we use larger filters we notice that the edge lines are thicker but there is lesser noise. That is, in the 3x3 filtered image, the lines are thinner compared to the 7x7 filtered image but it also has more noise comapred to the 7x7 image.

```python
# load clean cameraman image
cam_image_path = "./cameraman.png" # path of clean cameraman image
cam_img = cv2.imread(cam_image_path)
# converting to gray scale
cam_img = cv2.cvtColor(cam_img, cv2.COLOR_BGR2GRAY)

shape = np.shape(cam_img) 

#applying 3x3 filter
sobel3_x = cv2.Sobel(cam_img,cv2.CV_64F,1,0,ksize=3)
sobel3_y = cv2.Sobel(cam_img,cv2.CV_64F,0,1,ksize=3)
sobel3 = np.zeros(shape)
cv2.magnitude(sobel3_x, sobel3_y, sobel3)

#applying 5x5 filter
sobel5_x = cv2.Sobel(cam_img,cv2.CV_64F,1,0,ksize=5)
sobel5_y = cv2.Sobel(cam_img,cv2.CV_64F,0,1,ksize=5)
sobel5 = np.zeros(shape)
cv2.magnitude(sobel5_x, sobel5_y, sobel5)

#applying 7x7 filter
sobel7_x = cv2.Sobel(cam_img,cv2.CV_64F,1,0,ksize=7)
sobel7_y = cv2.Sobel(cam_img,cv2.CV_64F,0,1,ksize=7)
sobel7 = np.zeros(shape)
cv2.magnitude(sobel7_x, sobel7_y, sobel7)

#thresholding 10%
ret,sobel3_10 = cv2.threshold(sobel3,0.1*np.amax(sobel3),255,cv2.THRESH_BINARY)

ret,sobel5_10 = cv2.threshold(sobel5,0.1*np.amax(sobel5),255,cv2.THRESH_BINARY)

ret,sobel7_10 = cv2.threshold(sobel7,0.1*np.amax(sobel7),255,cv2.THRESH_BINARY)

#thresholding 20%
ret,sobel3_20 = cv2.threshold(sobel3,0.2*np.amax(sobel3),255,cv2.THRESH_BINARY)

ret,sobel5_20 = cv2.threshold(sobel5,0.2*np.amax(sobel5),255,cv2.THRESH_BINARY)

ret,sobel7_20 = cv2.threshold(sobel7,0.2*np.amax(sobel7),255,cv2.THRESH_BINARY)

#plot
titles = ['Sobel 3x3 10%', 'Sobel 3x3 20%', 
          'Sobel 5x5 10%', 'Sobel 5x5 20%',
          'Sobel 7x7 10%', 'Sobel 7x7 20%']
images = [sobel3_10, sobel3_20, 
          sobel5_10, sobel5_20, 
          sobel7_10, sobel7_20]
plt.subplot(1,2,1),plt.imshow(cam_img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
fig=plt.figure(figsize=(20, 24), dpi= 80)
for i in range(6):
    plt.subplot(3,2,i+1), plt.imshow(images[i], cmap = 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    
plt.show()

```

### Sobel edge detector on noisy cameraman

```python
# load noisy cameraman image
noisy_cam_image_path = "./gaussian_cameraman.png" # path of noisy cameraman image
noisy_cam_img = cv2.imread(noisy_cam_image_path)
# converting to gray scale
noisy_cam_img = cv2.cvtColor(noisy_cam_img, cv2.COLOR_BGR2GRAY)

shape = np.shape(noisy_cam_img)

# view image
plt.subplot(121), plt.imshow(noisy_cam_img, cmap = 'gray')
plt.title("Original "), plt.xticks([]), plt.yticks([])
plt.show()

#applying 3x3 sobel filter
sobel3_x = cv2.Sobel(noisy_cam_img,cv2.CV_64F,1,0,ksize=3)
sobel3_y = cv2.Sobel(noisy_cam_img,cv2.CV_64F,0,1,ksize=3)
sobel3 = np.zeros(shape)
cv2.magnitude(sobel3_x, sobel3_y, sobel3)

#thresholding 10%
ret,sobel3_10 = cv2.threshold(sobel3,0.1*np.amax(sobel3),255,cv2.THRESH_BINARY)

#thresholding 20%
ret,sobel3_20 = cv2.threshold(sobel3,0.2*np.amax(sobel3),255,cv2.THRESH_BINARY)

#plot
titles = ['Sobel 3x3 10%', 'Sobel 3x3 x 20%']
images = [sobel3_10, sobel3_20]
fig=plt.figure(figsize=(18, 16), dpi= 80)
for i in range(2):
    plt.subplot(1,2,i+1), plt.imshow(images[i], cmap = 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    

plt.show()
```

### Denoise noisy cameraman with box filter and then apply sobel edge detection

```python
# create a box filter
kernel = np.ones((3,3), np.float32)/(3*3)

# filter the image
clean_cam_img = cv2.filter2D(noisy_cam_img, -1, kernel)
shape = np.shape(clean_cam_img)

# view image
plt.subplot(121), plt.imshow(clean_cam_img, cmap = 'gray')
plt.title("Box filtered image "), plt.xticks([]), plt.yticks([])
plt.show()

#applying 3x3 sobel filter
sobel3_x = cv2.Sobel(clean_cam_img,cv2.CV_64F,1,0,ksize=3)
sobel3_y = cv2.Sobel(clean_cam_img,cv2.CV_64F,0,1,ksize=3)
sobel3 = np.zeros(shape)
cv2.magnitude(sobel3_x, sobel3_y, sobel3)

#thresholding 10%
ret,sobel3_10 = cv2.threshold(sobel3,0.1*np.amax(sobel3),255,cv2.THRESH_BINARY)

#thresholding 20%
ret,sobel3_20 = cv2.threshold(sobel3,0.2*np.amax(sobel3),255,cv2.THRESH_BINARY)

#plot
titles = ['Sobel 3x3 10%', 'Sobel 3x3 20%']
images = [sobel3_10, sobel3_20]
fig=plt.figure(figsize=(18, 16), dpi= 80)
for i in range(2):
    plt.subplot(1,2,i+1), plt.imshow(images[i], cmap = 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    

plt.show()
```

As can be seen from the above two cells, denoising before edge detection greatly reduces the noise in the final thresholded images. 

# Laplacian of Gaussian

### 1. Filter with 5x5 laplacian of gaussian kernel

```python
# load clean cameraman image
cam_image_path = "./cameraman.png" # path of clean cameraman image
cam_img = cv2.imread(cam_image_path)


# apply gaussian filter
gaus_img = cv2.GaussianBlur(cam_img,(5,5),0)

# converting to gray scale
gray_img = cv2.cvtColor(gaus_img, cv2.COLOR_BGR2GRAY)


# apply laplace function
lap_img = cv2.Laplacian(gray_img, cv2.CV_64F, ksize=5)

# view image
plt.subplot(121), plt.imshow(lap_img, cmap = 'gray')
plt.title("LoG Image "), plt.xticks([]), plt.yticks([])
plt.show()
```

### 2. Own code to detect edges

```python
# gradient of laplacian
grad = np.abs(np.diff(lap_img))
grad = np.pad(grad, ((0, 0), (0, 1)), 'reflect') # to keep shape

# check sign change of the laplacian 
cross = np.diff(np.sign(lap_img))
cross = np.pad(cross, ((0, 0), (0, 1)), 'constant', constant_values=0) # to keep  shape

# only keep the gradients that crosses zero and find 10% threshold
thresh = np.absolute(grad).max() * 0.1
grad[cross == 0] = 0

edge_lap = np.zeros(lap_img.shape)

# Detect the edges by finding the gradients greater than threshold
edge_lap[grad > thresh] = 255

# view image
fig=plt.figure(figsize=(18, 16), dpi= 80)
plt.subplot(121), plt.imshow(edge_lap, cmap = 'gray')
plt.title("Zero crossing Edge detection"), plt.xticks([]), plt.yticks([])
plt.show()
```

# Derivative of Gaussian

### 1.

```python
# load noisy cameraman image
noisy_cam_image_path = "./gaussian_cameraman.png" # path of noisy cameraman image
noisy_cam_img = cv2.imread(noisy_cam_image_path, cv2.IMREAD_GRAYSCALE)

# view image
plt.subplot(121), plt.imshow(noisy_cam_img, cmap = 'gray')
plt.title("Original "), plt.xticks([]), plt.yticks([])
plt.show()


# gaussian kernel
temp = cv2.getGaussianKernel(7,-1)
Gauss = np.outer(temp, temp.transpose())


#apply sobel to kernel
Gauss3_x = cv2.Sobel(Gauss,cv2.CV_64F,1,0,ksize=3)
Gauss3_y = cv2.Sobel(Gauss,cv2.CV_64F,0,1,ksize=3)


#apply kernel to noisy image
new_img_x = cv2.filter2D(noisy_cam_img, cv2.CV_64F, Gauss3_x)
new_img_y = cv2.filter2D(noisy_cam_img, cv2.CV_64F, Gauss3_y)
new_img = cv2.magnitude(new_img_x, new_img_y)

#thresholding 10%
ret,new_img_10 = cv2.threshold(new_img,0.1*np.amax(new_img),255,cv2.THRESH_BINARY)

#thresholding 20%
ret,new_img_20 = cv2.threshold(new_img,0.2*np.amax(new_img),255,cv2.THRESH_BINARY)

#plot
titles = ['Threshold 10%', 'Threshold 20%']
images = [new_img_10, new_img_20]
fig=plt.figure(figsize=(18, 16), dpi= 80)
for i in range(2):
    plt.subplot(1,2,i+1), plt.imshow(images[i], cmap = 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    

plt.show()
```

### 2.

```python
#filter image with 5x5 gaussian filter 
filtered_img =  cv2.GaussianBlur(noisy_cam_img, (5,5),0)

#applying 3x3 sobel filter
sobel3_x = cv2.Sobel(filtered_img,cv2.CV_64F,1,0,ksize=3)
sobel3_y = cv2.Sobel(filtered_img,cv2.CV_64F,0,1,ksize=3)
sobel3 = np.zeros(shape)
cv2.magnitude(sobel3_x, sobel3_y, sobel3)

#thresholding 10%
ret,sobel3_10 = cv2.threshold(sobel3,0.1*np.amax(sobel3),255,cv2.THRESH_BINARY)

#thresholding 20%
ret,sobel3_20 = cv2.threshold(sobel3,0.2*np.amax(sobel3),255,cv2.THRESH_BINARY)


#plot
titles = ['Original', 'Filtered', '10% threshold', '20% threshold']
images = [noisy_cam_img, filtered_img, sobel3_10, sobel3_20]
fig=plt.figure(figsize=(18, 16), dpi= 80)
for i in range(4):
    plt.subplot(2,2,i+1), plt.imshow(images[i], cmap = 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
    

plt.show()
```

# Canny edge detection

```python
# load lily image
lily_image_path = "./yellowlily.png" # path of clean cameraman image
lily_img = cv2.imread(lily_image_path)

# convert from BGR to RGB for displaying
lily_img_rgb = cv2.cvtColor(lily_img, cv2.COLOR_BGR2RGB)

# canny edge detection
edges = cv2.Canny(lily_img,400,1150,7)

# view image
fig=plt.figure(figsize=(18, 16), dpi= 80)
plt.subplot(121), plt.imshow(lily_img_rgb)
plt.title("Yellow lily"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(edges, cmap = 'gray')
plt.title("Edges"), plt.xticks([]), plt.yticks([])
plt.show()
```

Reducing the min value causes more edges to be decected, but with higher min values we are able to detect just the lily's outter edges.

Reducing the max value cause edges other than the lily to be detected, but increasing the max value by alot cause no edges to be detected at all.

The kernel size does not seem to have a very large effect on the edge detection.
